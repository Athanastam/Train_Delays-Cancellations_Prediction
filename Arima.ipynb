```python
import pandas as pd
import matplotlib.pyplot as plt
# load the dataset
df=pd.read_csv('merged_df.csv')
df.head()
```

```python
df.index = df.date
```

```python
df.head()
```

```python
df['Total Delay'] = df['Stop:Departure delay'] + df['Stop:Arrival delay']
```

```python
df.head()
```

```python

filtered_delays = df[(df['Total Delay'] <= 30)
                                 ]
```

```python
delay_df = filtered_delays[['Total Delay', 'tmin', 'tmax', 'date', 'wdir', 'wspd', 'prcp', 'pres']]
```

```python
delay_df.info()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Convert the 'timestamp' column to datetime 
delay_df['date'] = pd.to_datetime(delay_df['date'])

# Set the 'timestamp' column as the index
delay_df.set_index('date', inplace=True)

daily_delays = delay_df.resample('D').mean()  # 'D' for daily resampling

# Plot the resampled data
plt.plot(daily_delays.index, daily_delays['Total Delay'])
plt.title('Total Delay (Daily Average)')
plt.xlabel('Date')
plt.ylabel('Total Delay')
plt.xticks(rotation=45)
plt.show()
```

```python
from sktime.utils.plotting import plot_series
daily_delays.plot(subplots = True)
```

```python
# Calculate rolling statistics (mean and std deviation)
rolling_mean = daily_delays['Total Delay'].rolling(window=12).mean()
rolling_std = daily_delays['Total Delay'].rolling(window=12).std()

# Plot original data, rolling mean, and rolling std
plt.plot(daily_delays['Total Delay'], label='Original')
plt.plot(rolling_mean, label='Rolling Mean', color='orange')
plt.plot(rolling_std, label='Rolling Std', color='red')
plt.legend(loc='best')
plt.title('Rolling Mean & Standard Deviation')
plt.show()
```

```python
from statsmodels.tsa.stattools import adfuller

# Perform Augmented Dickey-Fuller test
result = adfuller(daily_delays['Total Delay'])

# Extract test results
adf_statistic = result[0]
p_value = result[1]
used_lags = result[2]
n_observations = result[3]
critical_values = result[4]

print('ADF Statistic:', adf_statistic)
print('p-value:', p_value)
print('Critical Values:')
for key, value in critical_values.items():
    print(f'  {key}: {value}')

# Interpretation of p-value
if p_value < 0.05:
    print("Reject the null hypothesis - The series is stationary")
else:
    print("Fail to reject the null hypothesis - The series is non-stationary")
```

```python
# First difference
daily_delays['Total Delay Differenced'] = daily_delays['Total Delay'].diff()

# Plot differenced data
plt.plot(daily_delays['Total Delay Differenced'])
plt.title('Differenced Total Delay')
plt.show()

# Re-run the ADF test on the differenced series
result = adfuller(daily_delays['Total Delay Differenced'].dropna())
print('ADF Statistic for Differenced Series:', result[0])
print('p-value for Differenced Series:', result[1])

```

```python
import numpy as np
import pandas as pd
from statsmodels.tsa.statespace.sarimax import SARIMAX
from sktime.split import temporal_train_test_split
from sktime.performance_metrics.forecasting import MeanSquaredError
from sktime.utils.plotting import plot_series
import itertools
import warnings

warnings.filterwarnings("ignore")

# Prepare exogenous features
exogenous_factors = daily_delays[['tmin', 'tmax', 'wdir', 'wspd', 'prcp', 'pres']]

# Split into train and test sets
tst_size = 30
y = daily_delays['Total Delay Differenced']
y_train, y_test = temporal_train_test_split(y, test_size=tst_size)
X_train, X_test = temporal_train_test_split(exogenous_factors, test_size=tst_size)

y_train = y_train.asfreq('D')
X_train = X_train.asfreq('D')

# Parameter ranges to search
p = d = q = range(0, 3)
P = D = Q = range(0, 2)
s = 7  # Weekly seasonality for daily data

# Generate combinations
pdq = list(itertools.product(p, d, q))
seasonal_pdq = list(itertools.product(P, D, Q, [s]))

# Track best model
best_score = float("inf")
best_params = None
results_table = []

for param in pdq:
    for seasonal_param in seasonal_pdq:
        try:
            model = SARIMAX(endog=y_train,
                            exog=X_train,
                            order=param,
                            seasonal_order=seasonal_param,
                            enforce_stationarity=False,
                            enforce_invertibility=False)
            result = model.fit(disp=False)
            y_pred = result.predict(start=len(y_train), end=len(y_train)+tst_size-1, exog=X_test)
            rmse = MeanSquaredError(square_root=True)(y_test, y_pred)
            results_table.append((param, seasonal_param, rmse))

            if rmse < best_score:
                best_score = rmse
                best_params = (param, seasonal_param)
                print(f"New best RMSE: {rmse:.4f} with params {param} & seasonal {seasonal_param}")
        except Exception as e:
            continue

# Sort and display top results
results_df = pd.DataFrame(results_table, columns=['order', 'seasonal_order', 'RMSE'])
results_df = results_df.sort_values(by='RMSE')
print("\nTop 5 parameter combinations:")
print(results_df.head())

# Fit final model with best parameters
best_order, best_seasonal_order = best_params
final_model = SARIMAX(endog=y_train,
                      exog=X_train,
                      order=best_order,
                      seasonal_order=best_seasonal_order,
                      enforce_stationarity=False,
                      enforce_invertibility=False)

final_result = final_model.fit(disp=False)
y_pred = final_result.predict(start=len(y_train), end=len(y_train)+tst_size-1, exog=X_test)

# Plot results
plot_series(y_train, y_pred, y_test, labels=["y_train", "y_pred", "y_test"])
print(f"\nFinal model RMSE: {MeanSquaredError(square_root=True)(y_test, y_pred):.4f}")

```

```python
# Permutation Feature Importance for SARIMAX
from copy import deepcopy

baseline_rmse = MeanSquaredError(square_root=True)(y_test, y_pred)
feature_importance = {}

print(f"\nBaseline RMSE: {baseline_rmse:.4f}")

for col in X_test.columns:
    # Copy the test data to permute
    X_test_permuted = X_test.copy()
    X_test_permuted[col] = np.random.permutation(X_test_permuted[col].values)

    # Predict with permuted data
    y_pred_permuted = final_result.predict(
        start=len(y_train), end=len(y_train)+tst_size-1, exog=X_test_permuted
    )

    # Calculate RMSE difference
    permuted_rmse = MeanSquaredError(square_root=True)(y_test, y_pred_permuted)
    importance = permuted_rmse - baseline_rmse
    feature_importance[col] = importance

    print(f"{col}: ΔRMSE = {importance:.4f}")

# Convert to DataFrame
importances_df = pd.DataFrame.from_dict(feature_importance, orient='index', columns=['ΔRMSE'])
importances_df = importances_df.sort_values(by='ΔRMSE', ascending=False)

# Plot feature importance (optional)
importances_df.plot(kind='bar', legend=False, title='Permutation Feature Importance (SARIMAX)', ylabel='ΔRMSE', xlabel='Feature', figsize=(10, 5))
```

```python
# Make sure importances_df is sorted
importances_df = importances_df.sort_values(by='ΔRMSE', ascending=False)

# Create the plot
plt.figure(figsize=(10, 5))
importances_df.plot(kind='bar', legend=False)
plt.title('Permutation Feature Importance (SARIMAX)')
plt.ylabel('ΔRMSE')
plt.xlabel('Feature')
plt.tight_layout()

# Save as PNG
plt.savefig('sarimax_feature_importance.png', dpi=300)
plt.close() 
```
